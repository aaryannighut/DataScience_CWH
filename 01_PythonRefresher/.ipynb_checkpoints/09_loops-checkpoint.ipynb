# For Loop

```python
l = [3, 5, 6, 67]
```

```python
for item in l:
    print(item)
```

```python
for i in range(10):
    print(i)
```

```python
for i in range(1, 11):
    print(i)
```

# While Loop

```python
i = 0
while i<12:
    print(i)
    i = i + 1
```

 # Loop Control Statements

```python
for item in l:
    pass       # Dont do anything

print("Hello")
```

```python
l = ["apples", "bananas", "tomatoes"]
for item in l:
    if(item == "bananas"):
        break     # stop this loop right here 
    print(item)
```

```python
l = ["apples", "bananas", "tomatoes"]
for item in l:
    if(item == "bananas"):
        continue     # skip this particular iteration of the loop
    print(item)
```
