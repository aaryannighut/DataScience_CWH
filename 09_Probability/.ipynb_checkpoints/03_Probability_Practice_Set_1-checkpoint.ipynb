# Questions

## 1. You roll a fair six-sided die. What is the probability of getting an even number?

### Solution

**Sample Space**

```text
S = {1, 2, 3, 4, 5, 6}
```

**Even Numbers**

```text
E = {2, 4, 6}
```

Number of favorable outcomes = **3**

Total number of outcomes = **6**

P(Even Number) = 3 / 6 = 1 / 2 = 0.5

**Answer:** **1/2 (0.5 or 50%)**

---

# 2. Two Coins are Tossed. Find the Sample Space Associated with this Random Experiment.

### Solution

Each coin has two possible outcomes:

```text
H = Head
T = Tail
```

**Sample Space**

```text
S = {(HH), (HT), (TH), (TT)}
```

Number of possible outcomes:

```text
n(S) = 4
```

**Answer:**

```text
{(HH), (HT), (TH), (TT)}
```

---

# 3. Two Six-Sided Dice are Rolled. What is the Probability that the Sum is 7?

### Solution

Total possible outcomes:

6 * 6 = 36

The favorable outcomes are:

```text
(1,6)
(2,5)
(3,4)
(4,3)
(5,2)
(6,1)
```

Number of favorable outcomes = **6**

Total outcomes = **36**

P(Sum=7) = 6 / 36 = 1 / 6

**Answer:** **1/6 (≈ 0.1667 or 16.67%)**

---

# 4. A Bag Contains 3 Red, 2 Blue, and 5 Green Balls. What is the Probability of Randomly Selecting a Red Ball?

### Solution

Total balls:

3+2+5=10

Number of red balls:

```text
3
```

P(Red) = 3 / 10 = 0.3

**Answer:** **3/10 (0.3 or 30%)**

---

# 5. A Card is Drawn at Random from a Standard Deck. Find the Probability of:

A standard deck contains **52 cards**.

---

## (a) Getting a Heart

There are **13 Hearts** in the deck.

P(Heart) = 13 / 52 = 1 / 4

**Answer:** **1/4 (0.25 or 25%)**

---

## (b) Getting a Face Card

Face cards are:

- Jack (J)
- Queen (Q)
- King (K)

Each suit has **3 face cards**.

There are **4 suits**.

Total face cards:
 
3 * 4=12

P(Face Card) = 12 / 52 = 3 / 13

**Answer:** **3/13 (≈ 0.2308 or 23.08%)**
