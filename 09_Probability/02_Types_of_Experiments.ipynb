# Types of Experiments in Probability

---

# Types of Experiments

## 1. Deterministic Experiment

A **Deterministic Experiment** always produces the **same outcome** when repeated under **identical conditions**.

### Example

- Turning on a working electric bulb using a functioning switch.
- Water boils at **100°C** under standard atmospheric pressure.

---

## 2. Probabilistic (Random) Experiment

A **Probabilistic (Random) Experiment** has an **uncertain outcome** and may produce different results even when repeated under identical conditions.

### Example

- Tossing a coin.
- Rolling a die.
- Drawing a card from a deck.

---

# Sample Space

The **Sample Space** is the **set of all possible outcomes** of a probabilistic experiment.

It is usually denoted by **S**.

---

# Number of Possible Outcomes

## For Coins

The number of possible outcomes is:

\[
2^n
\]

Where:

- **n** = Number of coins

### Examples

| **Number of Coins (n)** | **Possible Outcomes** |
|--------------------------|----------------------:|
| 1 | \(2^1 = 2\) |
| 2 | \(2^2 = 4\) |
| 3 | \(2^3 = 8\) |
| 4 | \(2^4 = 16\) |

---

## For Dice

The number of possible outcomes is:

\[
6^n
\]

Where:

- **n** = Number of dice

### Examples

| **Number of Dice (n)** | **Possible Outcomes** |
|-------------------------|----------------------:|
| 1 | \(6^1 = 6\) |
| 2 | \(6^2 = 36\) |
| 3 | \(6^3 = 216\) |
