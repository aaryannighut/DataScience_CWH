# Visualizing the Data

Before handling **missing values** or training models, it's important to **visualize the data** to uncover patterns, relationships, and potential issues.

---

# Geographical Scatter Plot

Visualize the geographical distribution of the data.

```python
df.plot(
    kind="scatter",
    x="longitude",
    y="latitude",
    grid=True,
    alpha=0.2
)

plt.show()
```

- `alpha=0.2` makes **overlapping points more visible**.
- This helps reveal **data clusters** and **high-density areas**, such as coastal regions.

---

# Correlation Matrix

To understand relationships between **numerical features**, compute the **correlation matrix**.

```python
corr_matrix = df.corr()
```

Check how strongly each attribute correlates with the target.

```python
corr_matrix["median_house_value"].sort_values(ascending=False)
```

This helps identify **useful predictors**.
For example, **`median_income`** usually shows a **strong positive correlation** with **house prices**.

---

# Scatter Matrix

Plot selected features to see **pairwise relationships**.

```python
from pandas.plotting import scatter_matrix

attributes = [
    "median_house_value",
    "median_income",
    "total_rooms",
    "housing_median_age"
]

scatter_matrix(df[attributes], figsize=(12, 8))

plt.show()
```

This gives an overview of **which features are linearly related** and may be **good predictors**.

````python
import pandas as pd
import numpy as np
````

````python
df = pd.read_csv("housing.csv")
````

````python
df['income_cat'] = pd.cut(df['median_income'], bins=[0, 1.5, 3.0, 4.5, 6.0, np.inf], labels=[1, 2, 3, 4, 5])
````

````python
from sklearn.model_selection import StratifiedShuffleSplit
split = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42)
for train_index, test_index in split.split(df, df['income_cat']):
    strat_train_set = df.loc[train_index]
    strat_test_set = df.loc[test_index]
    
````

````python
# Lets remove the income_cat column
for sett in (strat_train_set, strat_test_set):
    sett.drop("income_cat", axis=1, inplace=True)
````

````python
df = strat_train_set.copy()
````

````python
df.plot(kind="scatter", x="latitude", y="longitude", grid=True)
````

````python
df.plot(kind="scatter", x="latitude", y="longitude", grid=True, alpha=0.2)
````

````python
df.plot(kind="scatter", x="latitude", y="longitude", grid=True, cmap="jet", c="median_house_value")
````

````python
df.drop(labels="ocean_proximity", inplace=True, axis=1)
df.corr()
````

````python
from pandas.plotting import scatter_matrix
attributes = ["housing_median_age", "median_income", "median_house_value"]
scatter_matrix(df[attributes], figsize=(12, 8))
````
