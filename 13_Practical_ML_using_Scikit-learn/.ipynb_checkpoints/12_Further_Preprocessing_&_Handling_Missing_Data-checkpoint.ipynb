# Handling Missing Data

Some features, like **`total_bedrooms`**, contain **missing values**.
You can:
- Drop rows with missing values.
- Drop the entire column.
- Impute missing values *(recommended).*

We'll use **Option 3** using **`SimpleImputer`** from **Scikit-Learn**, which allows consistent handling across all datasets *(training, testing, and new data).*

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy="median")

housing_num = housing.select_dtypes(include=[np.number])
imputer.fit(housing_num)
```

This computes the **median** for each numerical column and stores it in **`imputer.statistics_`**.

```python
>>> imputer.statistics_
array([-118.51 ,  34.26 ,  29. , 2125. , 434. , 1167. , 408. , 3.5385])
```

Now apply the learned medians to transform the data.

```python
X = imputer.transform(housing_num)
```

# Other Available Strategies

### `"mean"`
Replaces missing values with the **mean** value.

### `"most_frequent"`
Replaces missing values with the **most common value**.
Can also handle **categorical data**.

### `"constant"`
Fills missing values with a **fixed value** using the `fill_value` parameter.
Example:

```python
SimpleImputer(strategy="constant", fill_value=0)
```
