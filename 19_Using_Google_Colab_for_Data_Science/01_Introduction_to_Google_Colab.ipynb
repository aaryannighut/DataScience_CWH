# Introduction to Google Colab

---

Google Colab is one of the most underrated tools ever, trust me. The whole idea is that it provides a runtime you can use anywhere: your phone, your iPad/ tablet, or your laptop.

It supports Python, Julia, and R, but we are mostly interested in Python.

To get started:

Go to your Google Drive and create a new Colab notebook.

---

# Common Questions

### Q: If we already have Jupyter, why should I use Colab?

**A:** Because Colab lets you run multiple Python versions and runtimes in parallel, without messing up your local setup. You also don’t have to install anything — just open and run.

---

### Q: Can I use it with AI?

**A:** Yes! Colab has Gemini integrated directly into the notebook environment. This is a complete gamechanger for running AI workflows, generating code, debugging, or automating tasks inside the notebook.

---

### Q: Is it free or paid?

**A:** For most users, the free version is more than enough. Paid tiers are useful only if you need heavy GPUs or very long session times.

---

### Q: How long can I use it?

**A:** According to the official FAQ:

> “In the free version, notebooks can run for up to 12 hours, depending on availability and usage patterns.”

It also mentions:

> “Runtimes will time out if you stay idle.”

So keep your notebook active and running!

---

### Q: What if I need it to read or process files?

**A:** You can easily:

- Mount your Google Drive directly into the notebook
- Upload files into the runtime
- Download output files from the notebook

We’ll explore these features in detail later.
