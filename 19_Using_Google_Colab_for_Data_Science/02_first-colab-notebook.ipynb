# Creating our First Colab Notebook

---

The first thing you should do in Colab is rename your notebook so it’s easy to find later. Just click on the title at the top and give it a meaningful name.

# Playground Mode

Colab also offers Playground Mode, which is basically a safe sandbox environment where you can experiment freely without affecting your original notebook. Perfect for testing code, trying ideas, or learning new concepts.

---

# Runtime Behavior

Colab automatically manages system resources. One thing to keep in mind: Expect Colab to clear your RAM if your notebook stays idle for around 9 minutes. This is normal for the free tier.

# Plotting Charts

Colab supports all major visualization libraries like matplotlib, seaborn, and plotly. Just write your plotting code and the chart will render inline:

```python
import matplotlib.pyplot as plt
 
plt.plot([1, 2, 3], [10, 20, 30])
plt.title("Sample Chart")
plt.show()
```

# Changing GPUs and TPUs

A big advantage of Colab is easy hardware switching:

**Runtime → Change runtime type → Hardware accelerator**

You can choose:

- CPU
- GPU
- TPU

This is extremely useful for deep learning, model experimentation, and training tasks.

````python
import pandas as pd

# Create a sample DataFrame
data = {'Column A': [1, 2, 3, 4], 'Column B': ['apple', 'banana', 'orange', 'grape']}
df = pd.DataFrame(data)

# Define the filename for the Excel file
excel_filename = 'sample_data.xlsx'

# Save the DataFrame to an Excel file
df.to_excel(excel_filename, index=False)

print(f"Excel file '{excel_filename}' generated successfully!")
````
