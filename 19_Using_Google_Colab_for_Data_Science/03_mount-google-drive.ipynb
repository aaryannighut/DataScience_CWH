# Mounting drive to Google Colab

---

To access your files directly inside Colab, you can mount your Google Drive. This allows you to read/write files just like a regular filesystem.

---

# Mount Drive

```python
from google.colab import drive
drive.mount("/content/drive")
```

When you run this, Colab will ask for authorization. Once approved, your Drive will be available at:

```text
/content/drive
```

You'll typically see folders like:

- MyDrive
- SharedDrives

---

# Unmounting Drive

If you ever need to unmount the drive:

```python
drive.flush_and_unmount()
```

---
