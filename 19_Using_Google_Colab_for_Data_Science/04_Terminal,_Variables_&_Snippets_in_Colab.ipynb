# Terminal, Variables & Snippets in Colab

---

Google Colab provides powerful tools beyond regular Python execution. You can use a full Linux terminal, inspect your session variables, and securely store/use secrets.

---

# Using the Terminal in Google Colab

Colab runs on a remote Linux machine. You can access the terminal in two ways:

## 1. Using `!` or `%system` in a cell

You can run shell commands directly:

```python
!ls
!pwd
!nvidia-smi
!pip install numpy
```

---

## 2. Opening an Interactive Terminal

Colab also supports an interactive terminal:

- Click on **"Terminal"** button on the bottom left

This opens a full Linux shell inside the browser.

---

# What Can You Do in the Terminal?

You can:

- Install packages via `apt-get` or `pip`
- Manage files and directories
- View hardware (GPU/TPU) info
- Run long-running scripts
- Use Git to clone/pull/push repositories
- Run shell-based workflows (build systems, scripts, etc.)

Example commands:

```bash
git clone https://github.com/CodeWithHarry/MongoDB-Crash-Course-2026.git
sudo apt-get update
top
python your_script.py
```

---

# Viewing Variables in Colab

Colab provides a Variable Inspector:

Click on the **Variables** towards the lower left to open it!

This allows you to see:

- All Python variables created in the notebook
- Their values
- Types
- Sizes (for arrays, lists, tensors, etc.)

It’s great for debugging or exploring data.

You can also manually inspect variables:

```python
globals()
locals()
```

And for large variables:

```python
import sys
sys.getsizeof(my_large_list)
```

---

# Handling Secrets in Google Colab

Colab offers built-in integrations for securely using secrets which can be accessed by clicking the **key icon** towards the left.

You can use the UI to add/delete/copy and provide notebook the access to a secret.
