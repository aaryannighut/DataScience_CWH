# Running Whisper on Google Colab

---

Google Colab makes it extremely easy to run OpenAI Whisper for speech-to-text, translation, and audio processing; all without installing anything locally.

---

# 1. Install Whisper

```python
!pip install -U openai-whisper
```

---

# 2. Uploading an Audio File (Drag & Drop)

In Colab, you can simply drag and drop your audio file into the left sidebar’s **Files** panel. Once dropped, the file appears immediately inside `/content/`.

For example, if you drag a file named `audio.mp3`, it will be accessible like this:

```text
/content/audio.mp3
```

No extra code is required.

---

# 3. Running Whisper

```python
import whisper
 
model = whisper.load_model("large")
result = model.transcribe("/content/audio.mp3")  # Use the filename you dropped
print(result["text"])
```

---

# 4. Translating Non-English Audio

```python
result = model.transcribe("/content/audio.mp3", task="translate")
print(result["text"])
```

---

# 5. Reading the Final Output

```python
result["text"]
```
